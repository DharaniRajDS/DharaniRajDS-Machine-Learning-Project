**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from xgboost import XGBRegressor
```

**Importing the Datasets**

```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataset
```

```python
# Transforming Dataset into DataFrame using Pandas
house_price_dataframe = pd.DataFrame(house_price_dataset.data)
```

```python
house_price_dataframe.head()
```

```python
# Adding columns name to the DataFrame
house_price_dataframe = pd.DataFrame(house_price_dataset.data, columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe.head()
```

```python
# Adding price column to the Dataset
house_price_dataframe['price'] = house_price_dataset.target
```

```python
house_price_dataframe.head()
```

```python
# Finding the Shape of the DataFrame
house_price_dataframe.shape
```

```python
# Finding any Null value present int the DataFrame
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.dtypes
```

```python
# Summary Statistics
house_price_dataframe.describe()
```

**Understanding the Correlation between variables present in the Dataset**

1. Positive Correlation
2. Negative Correlation

```python
correlation = house_price_dataframe.corr()
```

```python
# Constructing a heatmap to understand the correlation between the variables
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, cbar=True, square=True, fmt='0.1f', annot=True, annot_kws={'size': 8}, cmap='coolwarm')
```

**Splitting the Data & Target**

```python
X = house_price_dataframe.drop(['price'], axis=1)
Y = house_price_dataframe['price']
```

```python
print(X)
print(Y)
```

**Splitting the Data into Training Data & Testing Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Model Training**

**XGBoost Regressor**

```python
# Loading the Model
model = XGBRegressor()
```

```python
# Training the Model with X_train
# model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test)
model.fit(X_train, Y_train)
```

**Evaluation**

**Prediction on Training Data**

```python
# Accuracy for Prediction on Training Data
#training_data_prediction = classifier.predict(X_train)
#training_data_accuracy = accuracy_score(training_data_prediction, Y_train)
#print('Accuracy score of the training data : ', training_data_accuracy)
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R Squared Error
error_score_1 = metrics.r2_score(Y_train, training_data_prediction)
print('R squared error : ', error_score_1)
# Mean Absolute Error
error_score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)
print('Mean Absolute Error : ', error_score_2)
```

If error score is 0 that is perfect accuracy. if 5-10 we have to adjust the model

**Visualizing the Actual Price & Predicted Price**

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices VS Predicted Prices')
plt.show()
```

**Prediction on Test Data**

```python
# Accuracy for Prediction on Training Data
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

```python
# R Squared Error
error_score_1 = metrics.r2_score(Y_test, test_data_prediction)
print('R squared error : ', error_score_1)
# Mean Absolute Error
error_score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)
print('Mean Absolute Error : ', error_score_2)
```
