# Importing the Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

# Data Collection & Processing

```python
# loading the data from CSV file to Pandas Dataframe
car_dataset = pd.read_csv('/content/car data.csv')
```

```python
# Inspecting the first 5 rows
car_dataset.head()
```

```python
# Checking the number of rows & columns
car_dataset.shape
```

```python
# getting some information about the dataset
car_dataset.info()
```

```python
# checking the number of missing values
car_dataset.isnull().sum()
```

```python
car_dataset.head()
```

```python
# checking the distribution of categorical variables
print(car_dataset['Fuel_Type'].value_counts())
print(car_dataset['Seller_Type'].value_counts())
print(car_dataset['Transmission'].value_counts())
```

# Encoding the Categorical Data

```python
# encoding the fuel type column
car_dataset.replace({'Fuel_Type': {'Petrol':0, 'Diesel':1, 'CNG':2}}, inplace=True)
```

```python
# encoding the seller type column
car_dataset.replace({'Seller_Type': {'Dealer':0, 'Individual':1}}, inplace=True)
```

```python
# encoding the transmission column
car_dataset.replace({'Transmission': {'Manual':0,'Automatic':1}}, inplace=True)
```

```python
car_dataset.head()
```

# Splitting the Data & Target

```python
X = car_dataset.drop(columns=['Car_Name', 'Selling_Price'], axis=1)
Y = car_dataset['Selling_Price']
```

```python
print(X)
print(Y)
```

# Splitting the Training & Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=2)
```

# Model Training -> Linear Regression

```python
# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
# training the linear regression model
lin_reg_model.fit(X_train, Y_train)
```

# Model Evaluation

```python
# Prediction on Training data
training_data_prediction = lin_reg_model.predict(X_train)

# R squared error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print('R squared error : ', error_score)
```

```python
# Prediction on Test data
test_data_prediction = lin_reg_model.predict(X_test)

# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print('R squared error : ', error_score)
```

```python
# Visualize the Actual Prices & Predicted Prices
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```

# Model Training -> Lasso

```python
# loading the lasso model
lasso_model = Lasso()

# training the lasso model
lasso_model.fit(X_train, Y_train)
```

# Model Evaluation

```python
# Prediction on training data
training_data_prediction = lasso_model.predict(X_train)

# R squared error
error_score = metrics.r2_score(Y_train, training_data_prediction)
print('R squared error : ', error_score)
```

```python
# Prediction on test data
test_data_prediction = lasso_model.predict(X_test)

# R squared error
error_score = metrics.r2_score(Y_test, test_data_prediction)
print('R squared error : ', error_score)
```

```python
# Visualize the Actual Price & Predicted Price
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices vs Predicted Prices')
plt.show()
```
