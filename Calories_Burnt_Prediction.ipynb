# Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

# Data Collection & Pre-Processing

```python
# loading the data from csv files to a pandas dataframe
calories = pd.read_csv('/content/calories.csv')
```

```python
# printing the first 5 rows of the dataframe
calories.head()
```

```python
# loading the data from csv files to a pandas dataframe
exercise = pd.read_csv('exercise.csv')
```

```python
# printing the first 5 rows of the dataframe
exercise.head()
```

# Combining two DataFrames

```python
calorie_data =pd.concat([exercise, calories['Calories']], axis=1)
```

```python
calorie_data.head()
```

```python
# checking the number of rows & columns
calorie_data.shape
```

```python
# getting some information about the dataset
calorie_data.info()
```

```python
# checking for missing values
calorie_data.isnull().sum()
```

# Data Analysis

```python
# get some statistical measures about the data
calorie_data.describe()
```

Positive Correlation

Negative Correlation

```python
# Calculate the correlation matrix, excluding non-numeric columns
correlation = calorie_data.select_dtypes(include=np.number).corr()

# Display the correlation matrix
print(correlation) # Or correlation
```

# converting the text data to numerical values

```python
calorie_data.replace({'Gender':{'male':0, 'female':1}}, inplace=True)
```

```python
calorie_data.head(2)
```

# Separating the Features & Target

```python
X = calorie_data.drop(columns=['User_ID', 'Calories'], axis=1)
Y = calorie_data['Calories']
```

```python
print(X)
print(Y)
```

# Train Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training -> XGBoost Regressor

```python
# loading the model
model = XGBRegressor()
```

```python
# training the model
model.fit(X_train, Y_train)
```

# Model Evaluation -> Mean Absolute Error

Prediction on Test Data

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print('Mean Absolute Error:', mae)
```
